```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Downloads/janestreet.csv')
data
```

```python
data.columns
```

```python
col = ['IBM', 'HTC', 'AAPL']
row = ['CREST', 'DTC', 'Strate']

matrix = np.reshape((100, 100, 100, 100, 100, 100, 100, 100, 100), (3, 3))
matrix2 = np.reshape((0, 0, 0, 0, 0, 0, 0, 0, 0), (3, 3))

row1 = ['Strate']
d1Strate = [100, 100, 100] #keep track of strate for 8/20
d2Strate = [0, 0, 0] #keep track of strate for 8/21

res = pd.DataFrame(matrix, columns=col, index=row)
res2 = pd.DataFrame(matrix2, columns=col, index=row)

res
```

```python
for i in range(len(data)):
    if data['Trade date'][i] == "8/20/18":
        
        if data['Counterparty'][i] != "BnL Corp.":
        
            if data['Security'][i] == "IBM":
            
                if data['Buy/Sell'][i] == 'B':
                    if data['CSD'][i] == 'CREST': 
                        res['IBM'][0] += data['Quantity'][i]
                    if data['CSD'][i] == 'DTC':
                        res['IBM'][1] += data['Quantity'][i]
                    if data['CSD'][i] == 'Strate':
                        d1Strate[0] += data['Quantity'][i]
                    
                elif data['Buy/Sell'][i] == 'S':
                    if data['CSD'][i] == 'CREST':
                        res['IBM'][0] -= data['Quantity'][i]
                    if data['CSD'][i] == 'DTC':
                        res['IBM'][1] -= data['Quantity'][i]
                    if data['CSD'][i] == 'Strate':
                        d1Strate[0] -= data['Quantity'][i]    
                    
            elif data['Security'][i] == "HTC":
                if data['Buy/Sell'][i] == 'B':
                    if data['CSD'][i] == 'CREST':
                        res['HTC'][0] += data['Quantity'][i]
                    if data['CSD'][i] == 'DTC':
                        res['HTC'][1] += data['Quantity'][i]
                    if data['CSD'][i] == 'Strate':
                        d1Strate[1] += data['Quantity'][i]    

                elif data['Buy/Sell'][i] == 'S':
                    if data['CSD'][i] == 'CREST':
                        res['HTC'][0] -= data['Quantity'][i]
                    if data['CSD'][i] == 'DTC':
                        res['HTC'][1] -= data['Quantity'][i]
                    if data['CSD'][i] == 'Strate':
                        d1Strate[1] -= data['Quantity'][i]

            elif data['Security'][i] == "AAPL":
                if data['Buy/Sell'][i] == 'B':
                    if data['CSD'][i] == 'CREST':
                        res['AAPL'][0] += data['Quantity'][i]
                    if data['CSD'][i] == 'DTC':
                        res['AAPL'][1] += data['Quantity'][i]
                    if data['CSD'][i] == 'Strate':
                        d1Strate[2] += data['Quantity'][i]

                elif data['Buy/Sell'][i] == 'S':
                    if data['CSD'][i] == 'CREST':
                        res['AAPL'][0] -= data['Quantity'][i]
                    if data['CSD'][i] == 'DTC':
                        res['AAPL'][1] -= data['Quantity'][i]
                    if data['CSD'][i] == 'Strate':
                        d1Strate[2] -= data['Quantity'][i]
```
